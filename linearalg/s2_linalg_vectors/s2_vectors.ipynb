```python
# Vector dot product
v1 = np.array([ 1, 2, 3, 4, 5 ])
v2 = np.array([ 0, -4,  -3, 6, 5 ])

# method 1
dp1 = sum(np.multiply(v1,v2))

# method 2
dp2 = np.dot(v1, v2)

# method 3
dp3 = np.matmul(v1, v2)

# method 4
dp4 = 0

for i in range(len(v1)):
    dp4 = dp4 + v1[i]*v2[i]
    
print(dp1, dp2, dp3, dp4)
```

```python
## Distributive property

# create random vectors
n = 10
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)

# the two results
res1 = np.dot( a , (b+c) )
res2 = np.dot(a,b) + np.dot(a,c)

# compare them
print(a)
print([ res1,res2 ])
```

```python
## Associative property

# create random vectors
n = 5
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)

# the two results
res1 = np.dot( a, np.dot(b,c))
res2 = np.dot( c, np.dot(a,b))

# compare them
# print(a, np.dot(b,c))
# print(c, np.dot(a,b))
print(res1)
print(res2)


### special cases where associative property works!
# 1) one vector is the zeros vector
# 2) a==b==c
```

```python
# vector length
v1 = np.array([1, 2, 3, 4, 5, 6])

vl1 = np.sqrt(np.dot(v1, v1))

vl2 = np.linalg.norm(v1)

print(vl1, vl2)
```

```python
# The dot product from a geometric perspective
v1 = np.array([ 2,  4, -3 ])
v2 = np.array([ 0, -3, -3 ])

# compute the angle between two vectors
# 1 rad = 180°/π（about 57.29577951°）
ang = np.arccos(np.dot(v1, v2) / (np.linalg.norm(v1)*np.linalg.norm(v2)))
print(ang)

# 3D
fig = plt.figure(figsize=plt.figaspect(1))
ax = fig.gca(projection='3d')
ax.plot([0, v1[0]],[0, v1[1]],[0, v1[2]],'b', linewidth=3)
ax.plot([0, v2[0]],[0, v2[1]],[0, v2[2]],'r', linewidth=3)

# make the plot look nicer
ax.plot([0, 0],[0, 0],[-4, 4],'k--')
ax.plot([0, 0],[-4, 4],[0, 0],'k--')
ax.plot([-4, 4],[0, 0],[0, 0],'k--')
plt.title('Angle between vectors: %s rad.' %ang)
plt.show()
```

```python
## equivalence of algebraic and geometric dot product formulas
# two vectors
v1 = np.array([ 2,  4, -3 ])
v2 = np.array([ 0, -3, -3 ])

# algebraic
dp_a = np.dot( v1,v2 )

# geometric
dp_g = np.linalg.norm(v1)*np.linalg.norm(v2)*np.cos(ang)

print(dp_a, dp_g)
```

```python
# Vector Hadamard multiplication - element-wise 
# create vectors
w1 = [ 1, 3, 5 ]
w2 = [ 3, 4, 2 ]

w3 = np.multiply(w1,w2)
print(w3)
```

```python
# Vector outer product
v1 = np.array([  1, 2, 3 ])
v2 = np.array([ -1, 0, 1 ])

np.outer(v1, v2)
op = np.zeros((len(v1), len(v1)))

for i in range(0, len(v1)):
    for j in range(0, len(v2)):
        op[i,j] = v1[i] * v2[j]

print(op)
```

```python
# Vector cross product
v1  = [ -3,  2, 5 ]
v2  = [  4, -3, 0 ]

v3a = np.cross(v1, v2)

v3b = [ [v1[1]*v2[2] - v1[2]*v2[1]],
        [v1[2]*v2[0] - v1[0]*v2[2]],
        [v1[0]*v2[1] - v1[1]*v2[0]] ]

print(v3a, v3b)

fig = plt.figure()
ax = fig.gca(projection = '3d')

# draw plane
xx, yy = np.meshgrid(np.linspace(-10,10,10), np.linspace(-10,10,10))
z1 = (-v3a[0]*xx - v3a[1]*yy)/v3a[2]
ax.plot_surface(xx,yy,z1, alpha=.2)

ax.plot([0, v1[0]],[0, v1[1]],[0, v1[2]],'k')
ax.plot([0, v2[0]],[0, v2[1]],[0, v2[2]],'k')
ax.plot([0, v3a[0]],[0, v3a[1]],[0, v3a[2]],'r')

ax.view_init(azim=150,elev=45)
plt.show()
```
